```python
plt.plot(batsman['index'],batsman['V Kohli'],color='#bf0321') #graph for kohli
plt.plot(batsman['index'],batsman['RG Sharma'],color='#bf9234') #graph for sharma
plt.title('Rohit Sharma Vs Virat Kolhi Career Comparison') #used for giving title
plt.xlabel('Season') #used for naming the x axis
plt.ylabel('Runs Scored') #used for naming the yaxis
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='#bf0321',linestyle='dashed') #graph for kohli
#color='' used to give color / linesyles='' used to modifty the line dashed/solid/dashdot
plt.plot(batsman['index'],batsman['RG Sharma'],color='#bf9234',linestyle='solid') #graph for sharma
plt.title('Rohit Sharma Vs Virat Kolhi Career Comparison') #used for giving title
plt.xlabel('Season') #used for naming the x axis
plt.ylabel('Runs Scored') #used for naming the yaxis
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='#bf0321',linestyle='dashed', linewidth=5) #graph for kohli
#color='' used to give color / linesyles='' used to modifty the line dashed/solid/dashdot
#linewidth = it will describe
plt.plot(batsman['index'],batsman['RG Sharma'],color='#bf9234',linestyle='solid',linewidth=2) #graph for sharma
plt.title('Rohit Sharma Vs Virat Kolhi Career Comparison') #used for giving title
plt.xlabel('Season') #used for naming the x axis
plt.ylabel('Runs Scored') #used for naming the yaxis
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='#bf0321',linestyle='dashed', linewidth=2,marker='D',markersize=5,label='Virat') #graph for kohli
#color='' used to give color / linesyles='' used to modifty the line dashed/solid/dashdot
#linewidth = it will describe
plt.plot(batsman['index'],batsman['RG Sharma'],color='#000000',linestyle='solid',linewidth=2,marker='*',markersize=9,label="ROhit") #graph for sharma
plt.title('Rohit Sharma Vs Virat Kolhi Career Comparison') #used for giving title
plt.xlabel('Season') #used for naming the x axis
plt.ylabel('Runs Scored') #used for naming the yaxis
plt.legend(loc='best') #loc can be best center or any other mostly best is used
plt.grid() #to show the grid lines
plt.show()#to show the graph in other softwares
```
